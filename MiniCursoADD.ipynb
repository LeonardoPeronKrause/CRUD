```python
import os
import pandas as pd
import plotly.express as px

lista_arquivo = os.listdir("/content/drive/MyDrive/Mini Curso Python/Vendas")
display(lista_arquivo) # -> Display é um print que muda o jeito que as informações são mostradas
tabela_total = pd.DataFrame() # DataFrame é um tabela no python

for arquivo in lista_arquivo:
  if 'Vendas' in arquivo: # se o nome Vendas estiver no nome do arquivo
    tabela = pd.read_csv(f"/content/drive/MyDrive/Mini Curso Python/Vendas/{arquivo}")  # Importando o arquivo
    tabela_total = tabela_total.append(tabela) # O novo valor da tabela total é o antigo valor adicionado aos valores de 'tabela'

display(tabela_total) # Compilação de todos os arquivos do sistema

# Calcular o produto mais vendido
tabela_produtos = tabela_total.groupby('Produto').sum() # GROUPBY agrupa uma coluna que você passar dentro dos parêntes / .sum() soma os valores que correspondem ao Produto
tabela_produtos = tabela_produtos[['Quantidade Vendida', 'Preco Unitario']].sort_values(by='Quantidade Vendida', ascending=False) # seleciona apenas a tabela qnt vendida e preco unit / sort_values() vai ordenar valores / (by=...) voce informa qual coluna você quer / ascending= True ou False é igual a ordem crescente ou decrescente
display(tabela_produtos)

# Calcular o produto que mais faturou
tabela_total['Faturamento'] = tabela_total['Quantidade Vendida'] * tabela_total['Preco Unitario'] # Cria uma nova coluna, no caso a de Faturamento
tabela_faturamento = tabela_total.groupby('Produto').sum()
tabela_faturamento = tabela_faturamento[['Faturamento']].sort_values(by='Faturamento', ascending=False)
display(tabela_faturamento)

# Calcular a loja que mais vendeu em faturamento
tabela_lojas = tabela_total.groupby('Loja').sum()
tabela_lojas = tabela_lojas[['Faturamento']].sort_values(by='Faturamento', ascending=True)
display(tabela_lojas)

# Dashboard

grafico = px.bar(tabela_lojas, x=tabela_lojas.index, y='Faturamento')
grafico.show()














```
